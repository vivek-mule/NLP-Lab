#### Apply WordNet-based Word Sense Disambiguation to improve meaning interpretation in ambiguous sentences.

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
```

```python
sentence = "Earlier today I was playing with football on the ground."
word = "football"

tokens = word_tokenize(sentence)

sense = lesk(tokens, word)

print("Sentence:", sentence)
print("Ambiguous Word:", word)
print("Predicted Sense:", sense)
print("Definition:", sense.definition())
```
